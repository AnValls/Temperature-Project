```python
import os
import pandas as pd

print(os.getcwd())

os.chdir('C:\\Users\\Ana Valls\\Desktop\\CURSO BIG DATA\\IBM\\Python For Data Science\\Proyecto 2')  

df = pd.read_csv('temperaturas.csv')

print("Primeras filas del DataFrame:")
print(df.head())
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])  
df.dropna() 

print("\nInformación del DataFrame:")
print(df.info())
```

```python
print("\nResumen estadístico:")
print(df.describe())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df['Fecha'], df['Temperatura'], marker='o', linestyle='-', color='b')
plt.xlabel('Fecha')
plt.ylabel('Temperatura (°C)')
plt.title('Variación de Temperaturas')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

df['Mes'] = df['Fecha'].dt.month
df['Año'] = df['Fecha'].dt.year

temperaturas_mensuales = df.groupby('Mes')['Temperatura'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=temperaturas_mensuales, x='Mes', y='Temperatura', palette='coolwarm')
plt.xlabel('Mes')
plt.ylabel('Temperatura Media (°C)')
plt.title('Temperatura Media por Mes')
plt.show()

temperaturas_anuales = df.groupby('Año')['Temperatura'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=temperaturas_anuales, x='Año', y='Temperatura', marker='o')
plt.xlabel('Año')
plt.ylabel('Temperatura Media (°C)')
plt.title('Temperatura Media por Año')
plt.grid(True)
plt.show()
```
